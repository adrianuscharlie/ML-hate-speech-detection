```python
import pandas as pd
import re
import numpy as np
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
# set width  - 100
pd.options.display.max_colwidth=500
import urbandict
```

```python
data=pd.read_excel('./selenagomez.xlsx')
del data['Unnamed: 0']
data
```

```python
slang=pd.read_csv('./slang.csv',index_col=0)
slang={key:value for key,value in zip(slang['acronym'],slang['expansion'])}
slang
```

```python
data = data[data["text"].str.contains("RT") == False]
data
```

```python
#Lower text
def lowercase(text):
    return text.lower()
```

```python
#Remove unnecessary charracters
def removeUnnecessary(text):
    text=re.sub('\n','',text) # menghapus newline
    text=re.sub('rt','',text) # menghapus kata retweet
    text=re.sub(r'@\w+', '', text) # menghapus kata user
    text=re.sub(r"http\S+", "", text)
    text= re.sub('  +', ' ', text) # menghapus karakter ekstra
    text=re.sub('x.{3} | x.{2}','',text) # menghapus pola emoji 
    text=re.sub(r'\bim\b', 'I am', text) # im jadi I am
    text=re.sub(r'\bfyp\b', 'for you page', text, flags=re.IGNORECASE)# fyp jadi for you page
    text=re.sub(r'\big\b', 'instagram', text, flags=re.IGNORECASE) # ig jadi instagram
    return text
```

```python
#Remove non alphanumeric character
def removeNonAlphanumeric(text):
    text = re.sub('[^0-9a-zA-Z]+', ' ', text) 
    return text
```

```python
# Normalize word
from nltk.corpus import wordnet
def normalize(text):
    normalize_word=[]
    for word in text.split(' '):
        if word in slang.keys() and not wordnet.synsets(word):
            word=slang[word]
        normalize_word.append(word)
    return ' '.join(normalize_word)
```

```python
def removeStopWord(text):
    stop_words = set(stopwords.words("english"))
    word_tokens = word_tokenize(text)
    filtered_text = [word for word in word_tokens if word not in stop_words]
    return ' '.join(filtered_text)
```

```python
def lemmatizeWord(text):
    word_tokens = word_tokenize(text)
    # provide context i.e. part-of-speech
    lemmas = [lemmatizer.lemmatize(word, pos ='v') for word in word_tokens]
    return ' '.join(lemmas)
```

```python
def preProcessing(sentences):
    sentences=lowercase(sentences)
    sentences=removeUnnecessary(sentences)
    sentences=removeNonAlphanumeric(sentences)
    sentences=normalize(sentences)
    sentences=removeStopWord(sentences)
    sentences=lemmatizeWord(sentences)
    return sentences
```

```python
temp=data['text']
temp
```

```python
data['preprocessed']=data['text'].apply(preProcessing)
data
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
 
# function to print sentiments
# of the sentence.
def sentiment_scores(sentence):
 
    # Create a SentimentIntensityAnalyzer object.
    sid_obj = SentimentIntensityAnalyzer()
 
    # polarity_scores method of SentimentIntensityAnalyzer
    # object gives a sentiment dictionary.
    # which contains pos, neg, neu, and compound scores.
    sentiment_dict = sid_obj.polarity_scores(sentence)
     
 
    # # decide sentiment as positive, negative and neutral
    if sentiment_dict['compound'] >= 0.05 :
        return "Positive"
 
    elif sentiment_dict['compound'] <= - 0.05 :
        return "Negative"
 
    else :
        return "Neutral"
```

```python
data['sentiment']=data['preprocessed'].apply(sentiment_scores)
data['sentiment'].value_counts()
```
